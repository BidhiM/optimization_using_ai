```python
import numpy as np
f=lambda x: x**2+np.log(1+np.exp(-x**2))
g=lambda x: 2*x - (2*x*np.exp(-x**2))/(1 + np.exp(-x**2))
tol=0.1

def intervalsize(a0,b0):
    return b0-a0

def bisection(f,a0,b0):
    ss=intervalsize(a0,b0)
    while ss>tol:
        xk=(a0+b0)/2
        if g(xk) > 0:
            b0=xk
            ss=intervalsize(a0,xk)
        else:
            a0=xk
            ss=intervalsize(xk,b0)
    return xk

print("root",bisection(f,-2,2))
```
