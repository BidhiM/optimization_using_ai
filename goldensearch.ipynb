```python
import numpy as np
f=lambda x : x**2+np.log(1+np.exp(-x**2))
tol=0.1
rho=(3-np.sqrt(5))/2

def intervalsize(a0,b0):
    return b0-a0

def golden_search(f,a0,b0):
    ss=intervalsize(a0,b0)
    x1=a0+rho*ss
    x2=b0-rho*ss
    while ss>tol:
        if f(x1)<f(x2):
            ss=intervalsize(a0,x2)
            b0=x2
            x2=x1
            x1=a0+rho*intervalsize(a0,b0)
        else:
            ss=intervalsize(x1,b0)
            a0=x1
            x1=x2
            x2=b0-rho*intervalsize(a0,b0)
    return (a0+b0)/2
print("root=",golden_search(f,-2,2))

            
            
    
```
