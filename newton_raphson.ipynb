```python
import numpy as np
import matplotlib.pyplot as plt

f=lambda x : x**3 - 24*x**2 + 35*x + 42
df=lambda x : 3*x**2 - 48*x + 35
ddf=lambda x : 6*x - 48
tol=0.00001

X=np.linspace(-2,5,1000)
plt.figure(figsize=(6,4))
plt.plot(X,f(X),color='blue')
plt.axhline(0,color='black',alpha=0.3)
plt.grid()

def newtonrap(x0):
    while ddf(x0)!=0:
        xk=x0-(df(x0)/ddf(x0))
        plt.plot(xk,f(xk),'ro',markersize=4)
        if abs(df(xk)) < tol : return xk
        else: x0=xk
    return xk

print("root",newtonrap(5))
plt.show()
```
