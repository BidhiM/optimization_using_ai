```python
import numpy as np
import matplotlib.pyplot as plt

f=lambda x,y: x**2 + y**2-5
df_x=lambda x,y: 2*x 
df_y=lambda x,y: 2*y
ddf_x=lambda x,y:2
ddf_yx=0
ddf_xy=0
ddf_y=lambda x,y:2

iteration=15
x0=2
y0=-1
xk=[x0]
yk=[y0]
alpha = 0.1

x=np.linspace(-5,5,400)
y=np.linspace(-5,5,400)
X,Y=np.meshgrid(x,y)
z=f(X,Y)
plt.figure(figsize=(6,6))
plt.contour(X,Y,z,levels=15)

def raw_newton(x0,y0):
    for i in range(iteration):
        det_hf=(ddf_x(x0,y0)*ddf_y(x0,y0))-(ddf_xy*ddf_yx)
        x_new=x0-alpha*((ddf_y(x0,y0)/det_hf)*df_x(x0,y0) - (ddf_yx/det_hf)*df_y(x0,y0))
        y_new=y0-alpha*((-ddf_xy/det_hf)*df_x(x0,y0) + (ddf_x(x0,y0)/det_hf)*df_y(x0,y0))
        xk.append(x_new)
        yk.append(y_new)
        x0=x_new
        y0=y_new

raw_newton(2,-1)
plt.plot(xk,yk,'r--')
plt.show()
print(xk,end=' ')
```
