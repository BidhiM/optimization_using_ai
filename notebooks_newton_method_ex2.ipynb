```python
import numpy as np
import matplotlib.pyplot as plt

f = lambda x,y: x**4 + y**4 + x*y
df_x = lambda x,y: 4*x**3 + y
df_y = lambda x,y: 4*y**3 + x
ddf_x = lambda x,y: 12*x**2
ddf_y = lambda x,y: 12*y**2
ddf_xy = 1
ddf_yx = 1

iteration=10
x0=3
y0=1
xk=[x0]
yk=[y0]
alpha = 0.5

x=np.linspace(-10,10,400)
y=np.linspace(-10,10,400)
X,Y=np.meshgrid(x,y)
z=f(X,Y)
plt.figure(figsize=(6,6))
plt.contour(X,Y,z,levels=15)

def raw_newton(x0,y0):
    for i in range(iteration):
        det_hf=(ddf_x(x0,y0)*ddf_y(x0,y0))-(ddf_xy*ddf_yx)
        x_new=x0-alpha*((ddf_y(x0,y0)/det_hf)*df_x(x0,y0) - (ddf_yx/det_hf)*df_y(x0,y0))
        y_new=y0-alpha*((-ddf_xy/det_hf)*df_x(x0,y0) + (ddf_x(x0,y0)/det_hf)*df_y(x0,y0))
        xk.append(x_new)
        yk.append(y_new)
        x0=x_new
        y0=y_new

raw_newton(3,1)
plt.plot(xk,yk,'r--')
plt.show()
print(xk,end=' ')
print(yk,end=' ')
```
