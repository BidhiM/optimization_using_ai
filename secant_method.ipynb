```python
import numpy as np
f=lambda x : x**3 -24*x**2 +35*x+42
df=lambda x: 3*x**2 - 48*x + 35
tol=0.00001

def secant(x0,x1):
    while df(x0) != df(x1) :
        x2=x1-(df(x1)*(x1-x0))/(df(x1)-df(x0))
        if abs(df(x2))<tol: return x2
        else:
            x0=x1
            x1=x2
    return x2

print("root is=",secant(4,5))
```

```python
import numpy as np
import matplotlib.pyplot as plt
f=lambda x : x**3 -24*x**2 +35*x+42
df=lambda x: 3*x**2 - 48*x + 35
tol=0.00001

X=np.linspace(-2,10,1000)
plt.figure(figsize=(6,4))
plt.grid()
plt.plot(X,f(X),color='blue')
plt.axhline(0,alpha=0.3)
plt.plot(X,df(X),color='green')

def secant(x0,x1):
    while df(x0) != df(x1) :
        x2=x1-(df(x1)*(x1-x0))/(df(x1)-df(x0))
        plt.plot(x2,f(x2),'ro',markersize=2)
        if abs(df(x2))<tol: return x2
        else:
            x0=x1
            x1=x2
            plt.plot(x0,df(x0),'ko',markersize=1.5)
            plt.plot(x1,df(x1),'ro',markersize=1.5)
    plt.plot(x2,df(x2),'ro',markersize=2.5)
    return x2


print("root is=",secant(4,5))
```
